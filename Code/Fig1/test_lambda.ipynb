```python
import numpy as np
import matplotlib.pyplot as plt
from lambda_system import UCCZS, lambda_system
```

```python
t = np.linspace(0, 1000, 1001)
Delta = np.linspace(-0.003, 0.003, 1001)

g = 0.00162
phi = np.pi
in1 = 1
in2 = 0
in3 = 0

Delta = 2 * np.pi * Delta
g = 2 * np.pi * g
norm = np.sqrt(in1 * np.conj(in1) + in2 * np.conj(in2) + in3 * np.conj(in3))
in1 = in1 / norm
in2 = in2 / norm
in3 = in3 / norm
```

```python
X, Y = np.meshgrid(t, Delta)

pop1 = UCCZS.P1(X, 0, Y, g, phi, in1, in2, in3)
pop2 = UCCZS.P2(X, 0, Y, g, phi, in1, in2, in3)
pop3 = UCCZS.P3(X, 0, Y, g, phi, in1, in2, in3)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(8, 16))
ax[0].contourf(X, Y / 2 / np.pi, np.real(pop1), levels=100, vmin=0.0, vmax=1.0)
ax[1].contourf(X, Y / 2 / np.pi, np.real(pop2), levels=100, vmin=0.0, vmax=1.0)
ax[2].contourf(X, Y / 2 / np.pi, np.real(pop3), levels=100, vmin=0.0, vmax=1.0)
# plt.title("Population 1")
# plt.xlabel("Time (ns)")
# plt.ylabel("Detuning (ns)")
# plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
plt.plot(t, pop1[500, :]*np.exp(-t/(7*1000)))
plt.plot(t, pop2[500, :]*np.exp(-t/(7*1000)))
plt.plot(t, pop3[500, :]*np.exp(-t/(7*1000)))
```
